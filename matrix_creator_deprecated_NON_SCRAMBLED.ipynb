```python
import os
import networkx as nx
import matplotlib.pyplot as plt

# Set the path to the folder containing the edgelists
folder_path = "/home/pher/Documents/deeplearning/complex_network/graph_generators/edgelists_ER/"

# Create the folder for the images if it doesn't exist
images_folder = os.path.join(folder_path, "images")
os.makedirs(images_folder, exist_ok=True)

# Set the pixel resolution of the resulting image
img_width = 800
img_height = 800

# Set the size of the nodes and the line width of the edges as global variables
node_size = 50
edge_width = 1

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    # Check if the file is an edgelist file
    if filename.startswith("edgelist_") and filename.endswith(".txt"):
        # Load the edgelist file into a graph
        graph = nx.read_edgelist(os.path.join(folder_path, filename))
        # Create a visualization of the graph with custom pixel resolution, node size, and edge width
        plt.figure(figsize=(img_width/100, img_height/100), frameon=False)
        pos = nx.spring_layout(graph, seed=42)
        nx.draw_networkx_nodes(graph, pos, node_size=node_size, node_color='r', alpha=0.5)
        nx.draw_networkx_edges(graph, pos, width=edge_width, alpha=0.5)
        #nx.draw_networkx_labels(graph, pos, font_color='w', font_size=10)
        # Save the visualization as an image in the images folder with custom pixel resolution and no bounding box
        image_filename = os.path.join(images_folder, filename[:-4] + ".png")
        plt.savefig(image_filename, dpi=100)
        # Close the current figure to free up memory
        plt.close()
```
